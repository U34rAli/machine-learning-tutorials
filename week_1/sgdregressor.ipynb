```python
import matplotlib.pyplot as plt
from sklearn import linear_model, datasets, model_selection
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import PolynomialFeatures

X, y = datasets.load_diabetes(return_X_y=True)
X_tain, X_test, y_tain, y_test = model_selection.train_test_split(X[:,None,2], y, random_state=42, test_size = 0.3)
```

```python
reg = linear_model.SGDRegressor(penalty='elasticnet', max_iter=10000, alpha=0.0001)
```

```python
reg.fit(X_tain, y_tain)
```

```python
y_pred = reg.predict(X_test)
```

```python
plt.scatter(X_tain, y_tain)
plt.plot(X_test, y_pred, linewidth=3, color='red')
```

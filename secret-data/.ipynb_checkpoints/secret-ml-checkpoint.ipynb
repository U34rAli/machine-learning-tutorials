```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import sort
from scipy import stats
from xgboost import XGBClassifier
from xgboost import plot_importance
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.metrics import accuracy_score, classification_report
from sklearn.feature_selection import SelectFromModel
from sklearn.naive_bayes import GaussianNB
sns.set()
plt.rcParams['figure.figsize'] = [10, 10]
%matplotlib inline
```

```python
data = pd.read_csv('secretdata.csv')
use_equal_distrubution = True
data[data['class'] == 1].sample()
```

```python
ax = sns.countplot(data['class'],label="class")       # M = 212, B = 357
data['class'].value_counts()
```

```python
data = pd.read_csv('secretdata.csv')
if use_equal_distrubution:
    neg = data[data['class'] != 1].sample(492)
    pos = data[data['class'] == 1]
    # data = data.drop(pos.sample(200).index)
    data = neg.append(pos)
```

```python
X, y = data.iloc[:,:-1],data.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=7)
```

```python
data.corr()
```

```python
model = XGBClassifier()
model.fit(X_train,y_train)
```

```python
print(model.feature_importances_)
plt.bar(range(len(model.feature_importances_)), model.feature_importances_)
plt.show()
```

```python
plot_importance(model)
plt.show()
```

```python
y_pred = model.predict(X_test)
target_names = ['0', '1']
print(classification_report(y_test, y_pred, target_names=target_names))
```
